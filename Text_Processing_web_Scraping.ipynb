```python
reviews = []
for i in range(2,300,1):
    url ="https://www.flipkart.com/pigeon-stovekraft-limited-handy-chopper-l-vegetable-fruit/product-reviews/itme381172ddd7f7?pid=CPRFJHZAUBTHGWV7&lid=LSTCPRFJHZAUBTHGWV7FCA0BD&marketplace=FLIPKART&page="+str(i)
    print(url)
    page = requests.get(url)
    soup = BeautifulSoup(page.text)
    for i in soup.find_all('div', class_ = 't-ZTKy'):
        reviews.append(i.text)
```

```python
len(reviews)
```

```python
reviews
```

```python
import pandas as pd 
import numpy as np
```

```python
dictionary = { "Reviews":reviews }
df = pd.DataFrame(dictionary)
df
```
